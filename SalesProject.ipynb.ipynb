```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dp=pd.read_csv('/content/sales_data.csv',encoding='iso-8859-1')
dp
```

```python
dp['Amount'].isnull()
```

```python
dp.drop(columns=['Status', 'unnamed1'], inplace=True)
```

```python
dp.dropna(inplace=True)
```

```python
dp.rename(columns={'Marital_Status':'Marital'},inplace=True)
```

```python
dp['Marital'] = dp['Marital'].map({0: 'Single', 1: 'Married'})
```

```python
dp['Gender'] = dp['Gender'].map({'F': 'Female', 'M': 'Male'})
```

```python
plt.figure(figsize=(5,3),facecolor='pink')
plt.gca().set_facecolor('gray')
g=dp['Gender'].value_counts()
plt.pie(g.values,labels=g.index,autopct='0.2f%%',colors = ['purple', 'green'],shadow='True')
plt.title('Gender Distribution',color='red')
plt.tight_layout()
```

```python
a = dp.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(5,3))
plt.gca().set_facecolor('#F9F0CC')
a.plot(kind='bar',color=['#6DE6D3', '#E9C2F5'],edgecolor='black')
plt.title('Total Sales Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()
```

```python
b= dp.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(6,4))
plt.gca().set_facecolor('#EBB35D')
b.plot(kind='bar',color='#E7DDFF',edgecolor='black')
plt.title('Total Sales Amount by Age Group',color='purple')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()
```

```python
c= dp.groupby('State')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(6,4))
plt.gca().set_facecolor('#A2E188')
c.plot(kind='barh',color='#BD8E8F',edgecolor='black')
plt.title('Total Sales Amount by State',color='brown')
plt.xlabel('Total Amount')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
c= dp.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(6,4))
plt.gca().set_facecolor('#EEE0A8')
c.plot(kind='bar',color='#060270',edgecolor='black')
plt.title('Total Sales Amount by Occupation',color='green')
plt.xlabel('Total Amount')
plt.ylabel('Occupation')
plt.tight_layout()
plt.show()
```

```python
pd.pivot_table(dp,values='Amount',index='Product_Category',columns='Gender',aggfunc='sum').plot(kind='bar', stacked=True,color=['#6DE6D3', '#E9C2F5'])
plt.title('Product Category vs Gender Spending',color='blue')
plt.gca().set_facecolor('#E28182')
plt.tight_layout()
plt.show()
```

```python
e= dp.groupby('Marital')['Amount'].sum()
plt.figure(figsize=(3,3))
e.plot(kind='pie', autopct='%1.1f%%', colors=['#77EEF2','#F4D16C'],shadow='True')
plt.title('Sales Distribution by Marital',color='blue')
plt.tight_layout()
plt.show()

```
